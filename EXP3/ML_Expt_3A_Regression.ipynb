```python
import tensorflow as tf
from tensorflow import keras
# print(tf.__version__)
# print(keras.__version__)
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from google.colab import files
```

```python
from sklearn.model_selection import train_test_split
```

```python
uploaded=files.upload()
filename = list(uploaded.keys())[0]
```

```python
#housing=pd.read_csv('F:\\housing.csv')  # The data set is read into housing variable as panda frame
housing=pd.read_csv(filename)  # The data set is read into housing variable as panda frame
housing
```

```python
housing.pop('ocean_proximity')

housing1=housing
housing1
```

```python

mx=housing1.max()
print(mx)
mn=housing1.min()
print('minimum values=\n',mn)
```

```python
# Min-Max normalization
hnorm=(((housing1-housing1.min())/(housing1.max()-housing1.min()))) #min-max
print('hnorm=\n', hnorm)
print(hnorm.isna())
print(hnorm.isna().sum())
#hnorm.pop('total_bedrooms')
#print('hnorm=\n', hnorm')
```

```python
hnorm.pop('total_bedrooms')
hnorm
```

```python
target = hnorm.pop('median_house_value')
target
```

```python
hnorm
```

```python
y=target.values
x=hnorm.values
y
x
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y,random_state=0) # seperating test and train samples with default split ratio
print('X_train', X_train)
print('X_test', X_test)
```

```python
model = keras.models.Sequential([
keras.layers.Dense(7, activation="relu",input_shape=(7,)),
keras.layers.Dense(32,activation="relu"),
keras.layers.Dense(1)
])

```

```python
model.summary()
```

```python
model.compile(loss="mse", optimizer="Adam",metrics=['accuracy'])
```

```python
history = model.fit(X_train, y_train, epochs=100,verbose=2)
```

```python
htest = model.evaluate(X_test, y_test,verbose=2)
```
